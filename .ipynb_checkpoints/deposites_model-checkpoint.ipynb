```python
import xlrd
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import sklearn
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.preprocessing import StandardScaler
from pandas.plotting import scatter_matrix
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
```

```python
deposites = pd.read_excel("DataSet.xlsx", nrows=96)
deposites.index = deposites["Дата"]
deposites.drop("Дата", axis=1, inplace=True)
deposites.describe()
```

```python
matplotlib_params = {
    'axes.titlesize':'70',
    'xtick.labelsize':'50',
    'ytick.labelsize':'50'}

matplotlib.rcParams.update(matplotlib_params)

deposites.hist(
    bins=15,
    figsize=(100,100)
)
```

```python
train, test = deposites.loc[:'2017-12-01'], deposites.loc['2017-12-01':]

train_y = train["Прирост вкладов физических лиц в рублях (млн руб)"]
train_X = train.drop("Прирост вкладов физических лиц в рублях (млн руб)", axis=1)

test_y = test["Прирост вкладов физических лиц в рублях (млн руб)"]
test_X = test.drop("Прирост вкладов физических лиц в рублях (млн руб)", axis=1)
```

```python
class AttribsAdder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X1 = X["Нефть марки Юралс, долл./барр"]*X["Курс доллара к рублю, руб"]
        X2 = X["Номинальная средняя з/п, руб"]/X["Индекс потребительских цен, ед."]
        columns_names = list(X) + ["Нефть марки Юралс, руб./барр", "Реальные доходы"]
        return pd.DataFrame(np.c_[X, X1, X2], columns=columns_names, index=X.index)
```

```python
class RowsDeltaAdder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        delta_X = X.diff()
        delta_X_column_names = ["delta_" + col_name for col_name in list(X)]
        columns_names = list(X) + delta_X_column_names
        return pd.DataFrame(np.c_[X, delta_X], columns=columns_names, index=X.index)
```

```python
class RowsDeleter(BaseEstimator, TransformerMixin):
    def __init__(self, nrows=1):
        self.nrows=nrows        
    def fit(self, X, y=None):
        return self
    def transform(self, X):        
        return X.iloc[self.nrows:]
```

```python
class ColsSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names=None):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        if self.attribute_names:
            return pd.DataFrame(X[self.attribute_names].values, columns=self.attribute_names)
```

```python
named_features_pipeline = Pipeline([ 
    ('attribs_adder', AttribsAdder()),
    ('delta', RowsDeltaAdder()),
    ('rowsdeleter', RowsDeleter()),
]) 
```

```python
all_features_named = named_features_pipeline.fit_transform(train.copy())
corr_matrix= all_features_named.corr()
print(corr_matrix["Прирост вкладов физических лиц в рублях (млн руб)"])
```

```python
selected_cols = [
    "Прирост вкладов физических лиц в рублях (млн руб)",
    "delta_М0, руб",
    "delta_М2, руб",
    "delta_Номинальная средняя з/п, руб",
    "delta_Реальные доходы"    
]
scatter_matrix(all_features_named[selected_cols], figsize=(12,8))
```

```python
model_cols = [
    "delta_М2, руб",   
]

X_final_pipeline = Pipeline([
    ('attribs_adder', AttribsAdder()),
    ('delta', RowsDeltaAdder()),
    ('rowsdeleter', RowsDeleter()),
    ('selector', ColsSelector(model_cols)),
    ('std_scaler', StandardScaler())
])

y_final_pipeline = Pipeline([
    ('rowsdeleter', RowsDeleter()),
])

train_X_prepared = X_final_pipeline.fit_transform(train_X)
train_y_prepared = y_final_pipeline.fit_transform(train_y)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(train_X_prepared, train_y_prepared)
```

```python
test_X_prepared = X_final_pipeline.fit_transform(test_X)
test_y_prepared = y_final_pipeline.fit_transform(test_y)

test_y_predictions = lin_reg.predict(test_X_prepared)

lin_mse = mean_squared_error(test_y_predictions, test_y_prepared)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
matplotlib_params = {
    'xtick.labelsize':'10',
    'ytick.labelsize':'10'
}

matplotlib.rcParams.update(matplotlib_params)

train_y_predictions = lin_reg.predict(train_X_prepared)
plt.plot(test_X_prepared, test_y_predictions, "r-")
plt.plot(test_X_prepared, test_y_prepared, "b.")
plt.plot(train_X_prepared, train_y_predictions, "r-")
plt.plot(train_X_prepared, train_y_prepared, "b.")
```

```python
def display_scores(scores):
    print(scores)
    print(scores.mean())
    print(scores.std())
```

```python
lin_reg_scores = cross_val_score(lin_reg, test_X_prepared, test_y_prepared, scoring="neg_mean_squared_error", cv=2)
lin_reg_rmse_scores = np.sqrt(-lin_reg_scores)
```

```python
display_scores(lin_reg_scores)
```
